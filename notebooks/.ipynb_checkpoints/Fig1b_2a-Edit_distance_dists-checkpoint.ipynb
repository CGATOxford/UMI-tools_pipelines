# Edit distance distributions

In this notebook we will explore the distribution of edit distances between various different methods for deduplication and compare these to a null derived by random sampling. We will produce the figures that are Figure 1b and Figure 3a for the SR data set and Figure S1.

```python
import CGAT.Database as DB
import CGAT.IOTools as IOTools
import os
import re
import pandas as pd
import numpy as np
% load_ext rpy2.ipython
%R library(ggplot2)
%R library(Hmisc)
```

The edit_distance distributions are in seperate tables for each sample and method, so we need to run a query for each table and concatentate the results, indexing by sample and method. The null distribution is in a seperate column of each table, but we would like it to look like another "method" so its distribution can be compared side by side. Thus for the "unique" method we will also pull back the null column, rename it and append it to the end of the dataframe. 

```python
edit_distance_tables = DB.fetch_DataFrame("SELECT name FROM sqlite_master WHERE type='table' AND name LIKE '%_edit_distance' ", "csvdb")
def getTable(x):
    x = x["name"].iloc[0]
    track, method = re.match("(.+)_dedup_(.+)_edit", x).groups()
    method = re.sub("^u","_u", method)
    statement = ('''SELECT %(method)s as count,
                           edit_distance, 
                           '%(track)s' as Track,
                           '%(method)s' as Method
                    FROM %(x)s
                    WHERE edit_distance != 'Single_UMI' ''')
    results = DB.fetch_DataFrame(statement %locals(), "csvdb")
    if method == "_unique":
        method = "_unique_null"
        results2 = DB.fetch_DataFrame(statement % locals(), "csvdb")
        results2["Method"] = "Null"
        results = pd.concat([results, results2])

    return results
edit_distances = edit_distance_tables.groupby("name").apply(getTable)
edit_distances.index = edit_distances.index.droplevel(1)
edit_distances[["Protein","Rep"]] = edit_distances["Track"].str.extract("(.+)_.+_(.+)")
edit_distances = edit_distances[edit_distances["edit_distance"] != 6]
```

Figure 1b demonstrates that simply taking unique UMIs leaves a distribution of edit distances that is substantially different to the Null expectation. So we only want to show the "unique" and the Null methods. Because the SR dataset has so many proteins and replicates, we also only show replicate 1 for the main figure (data for the rest is in the supplement).

```python
%%R -w 116 -h 75 -u mm -r 300 -i edit_distances

library(ggplot2)
edit_distances = subset(edit_distances, Method %in% c("_unique","Null") & Rep == "R1" )
g <- ggplot(edit_distances) +  
     aes(Method, count, fill=as.factor(edit_distance), colour=NULL) + 
     geom_bar(stat="identity", position="fill", lwd=0.4) + 
     scale_fill_manual(name="Edit distance", values = topo.colors(7, alpha=0.9)) + 
     scale_x_discrete(limits=c("_unique","Null"), labels=c("Observed","Null"), name = "") + 
     ylab("Fraction") + scale_color_manual(guide="none", values="grey80") + 
     theme_bw(base_size=8) + 
     theme(panel.grid.major=element_blank(), 
           panel.grid.minor=element_blank(), 
           axis.text.x = element_text(angle=90, hjust=1, vjust=0.5), 
           legend.position="top", 
           legend.key.size=unit(0.6, "lines"),
           aspect.ratio=2) + 
     facet_grid(~Protein)
ggsave("plots/Figure1b.svg", plot = g)
print (g)
```

## Network topologies

This section of the notebook deals with different network topologies, where a topology means that either a network is a single UMI (e.g. not a network), has a single hub that all other nodes are related to, or is more complex. This is output by umi_tools with the setting --further-stats.

Here we categorize further into those with a single UMI, and those with more, so that fig1d can be produced. We also output some numbers for use in the manuscript.

```python
topologies = DB.fetch_DataFrame("SELECT * FROM topologies ","csvdb")
topologies[["Protein","Replicate"]] = topologies.track.str.extract("(.+)-.+-(.+)")
topologies["category"] = topologies["category"].str.capitalize()

complexity_lookup = {"Single node": "Single UMI",
                     "Single hub": "Multiple UMI",
                     "Complex": "Multiple UMI"}
topologies["complexity"] = topologies["category"].apply(lambda x: complexity_lookup[x])
complexity = topologies.groupby(["Protein","Replicate","complexity"]).sum().reset_index()
```

```python
node_counts = DB.fetch_DataFrame("SELECT * FROM node_counts", "csvdb")
node_counts["category"] = pd.cut(node_counts.category,
                            bins=[0,1,2, max(node_counts.category)],
                            labels=["1","2","3+"])
node_counts = node_counts.groupby(["track","category"]).sum().reset_index()
node_counts[["Protein","Replicate"]] = node_counts.track.str.extract("(.+)-.+-(R.)")
node_counts.head()
```

```python
%%R -h 63 -w 63 -u mm -r 300 -i topologies
library(dplyr)
print(head(topologies))
g <- topologies %>% 
     subset(Replicate == "R1") %>%
     ggplot() +
     aes(x=Protein, y=count, fill=category, colour = "NA") +
     geom_bar(stat="identity", position="fill",  lwd=0.3) +
     scale_fill_manual(values = rev(topo.colors(3, alpha=0.9)), name = "")  +
     scale_color_manual(guide="none", values="grey80") + 
     theme_bw(base_size=8) +
     ylab("Fraction of positions") +
     theme(legend.position="bottom",
           panel.grid.major=element_blank(), 
           panel.grid.minor=element_blank(),
           axis.text.x=element_text(angle=90,hjust=1, vjust=0.5 ),
           aspect.ratio=1.2           ,
           axis.title.x = element_blank(),
           legend.key.size=unit(0.4,"line"))

print(g)
#ggsave("plots/figureS1c.svg")
```
